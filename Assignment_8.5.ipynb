```python
#5) For each group of four consecutive calendar months in s, find the date on which the
#highest value occurred.
```

```python
import numpy as np
import pandas as pd
from pandas.tseries.offsets import BDay
dt = pd.date_range(start='1/1/2015',end='12/31/2015', freq=BDay(), tz='UTC')
dt = pd.to_datetime(dt.date)
s = pd.Series(np.random.rand(len(dt)),index=dt)
df = pd.DataFrame({'Date':dt, 'Day':dt.weekday_name, 'Month':dt.month_name(), 'Val':s},index=dt)
df.head(10)
```

```python
df.resample('4M').max()
```
